### Generate images by **bicubic**

```python
import tensorlayer as tl
import scipy
```

```python
test_lr_path = '../data/train_set/LR'
test_lr_img_list = sorted(tl.files.load_file_list(path=test_lr_path, regx='.*.jpg', printable=False))
save_path = '../output/SRGAN2/test_bicubic'

imgs = tl.vis.read_images(test_lr_img_list, path=test_lr_path)
out = [scipy.misc.imresize(img, [img.shape[0]*2, img.shape[1]*2], interp='bicubic', mode=None) for img in imgs]
for i in range(len(out)):
    tl.vis.save_image(out[i], os.path.join(save_path, '{}'.format(test_lr_img_list[i])))
```

### Comparison by psnr

```python
import numpy as np
import tensorlayer as tl
```

```python
test_hr_img_path = '../data/train_set/HR'
gen_hr_img_path = '../output/SRGAN2/test_srgan'
def psnr(img1, img2):
    '''
    img1 and img2 are two 3-dimention images
    '''
    return 10*np.log10(255*255/(np.square(img1-img2).mean()))
```

```python
test_hr_list = sorted(tl.files.load_file_list(path=test_hr_img_path, regx='.*.jpg', printable=False))
test_gen_list = sorted(tl.files.load_file_list(path=gen_hr_img_path, regx='.*.jpg', printable=False))

test_hr_imgs = tl.vis.read_images(test_hr_list[20:200], path=test_hr_img_path)
test_gen_imgs = tl.vis.read_images(test_gen_list[20:200], path=gen_hr_img_path)

# mean psnr
np.mean([psnr(img1,img2) for img1, img2 in zip(test_hr_imgs, test_gen_imgs)])
```
