```python
import os, sys
import numpy as np
import cv2
```

```python
import tensorflow as tf
from keras.models import Sequential, Model
from keras.layers import Input, Activation, Add, Dropout, ZeroPadding2D, Flatten, Reshape
from keras.layers import BatchNormalization, LeakyReLU, PReLU, Conv2D, Dense
from keras.layers import UpSampling2D, Lambda
from keras.applications.vgg19 import VGG19
from keras.initializers import RandomNormal, Zeros
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator
import keras.backend as K
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def image_loader(path, endswith='.jpg', height=None, width=None):
    '''
    load image data
    if height and width are not specified, return list
    if specified, return array
    '''

    lr_path = os.path.join(path, 'LR')
    hr_path = os.path.join(path, 'HR')
    
    lr_img_list = sorted([img for img in os.listdir(lr_path) if img.endswith(endswith)])
    hr_img_list = sorted([img for img in os.listdir(hr_path) if img.endswith(endswith)])

    lr_imgs = [cv2.cvtColor(cv2.imread(os.path.join(lr_path, img)), cv2.COLOR_BGR2RGB) / 255. for img in lr_img_list]
    hr_imgs = [cv2.cvtColor(cv2.imread(os.path.join(hr_path, img)), cv2.COLOR_BGR2RGB) / 255. for img in hr_img_list]

    if height == None and width == None:
        return lr_imgs, hr_imgs
    
    size = (width, height)
    lr_imgs = [cv2.resize(img, dsize=size, interpolation=cv2.INTER_CUBIC) for img in lr_imgs]
    lr_imgs = np.concatenate([np.expand_dims(img, axis=0) for img in lr_imgs])

    hr_imgs = [cv2.resize(img, dsize=size, interpolation=cv2.INTER_CUBIC) for img in hr_imgs]
    hr_imgs = np.concatenate([np.expand_dims(img, axis=0) for img in hr_imgs])
    return lr_imgs, hr_imgs
```

```python
lr_image, hr_image = image_loader('../data/train_set', height=320, width=380)
```

```python
# residual block
def residual_block(model, kernal_size, filters, strides):
    '''
    
    '''
    gen = model
    
    model = Conv2D(filters = filters, kernel_size = kernal_size, strides = strides, padding = "same")(model)
    model = BatchNormalization(momentum = 0.5)(model)
    # Using Parametric ReLU
    model = PReLU(alpha_initializer='zeros', alpha_regularizer=None, alpha_constraint=None, shared_axes=[1,2])(model)
    model = Conv2D(filters = filters, kernel_size = kernal_size, strides = strides, padding = "same")(model)
    model = BatchNormalization(momentum = 0.5)(model)
        
    model = Add()([gen, model])
    
    return model

# upsampling block
def upsampling_block(model, kernal_size, filters, strides):
    '''
    
    '''
    model = Conv2D(filters = filters, kernel_size = kernal_size, strides = strides, padding = "same")(model)
    model = UpSampling2D(size = 2)(model)
    model = LeakyReLU(alpha = 0.2)(model)
    
    return model

def discriminator_block(model, filters, kernel_size, strides):
    
    model = Conv2D(filters = filters, kernel_size = kernel_size, strides = strides, padding = "same")(model)
    model = BatchNormalization(momentum = 0.5)(model)
    model = LeakyReLU(alpha = 0.2)(model)
    
    return model
```

```python
# generator
def Generator():
    '''
    
    '''
    gen_input = Input(shape=(None, None, 3))
    
    model = Conv2D(filters = 64, kernel_size = 9, strides = 1, padding = "same")(gen_input)
    model = PReLU(alpha_initializer='zeros', alpha_regularizer=None, alpha_constraint=None, shared_axes=[1,2])(model)
    
    gen_model = model
    
    # Using 16 Residual Blocks
    for index in range(1):
        model = residual_block(model, 3, 64, 1)
    
    model = Conv2D(filters = 64, kernel_size = 3, strides = 1, padding = "same")(model)
    model = BatchNormalization(momentum = 0.5)(model)
    model = Add()([gen_model, model])
    
    for index in range(1):
        model = upsampling_block(model, 3, 256, 1)
    
    model = Conv2D(filters = 3, kernel_size = 9, strides = 1, padding = "same")(model)
    model = Activation('tanh')(model)
    
    generator_model = Model(inputs=gen_input, outputs=model)
    return generator_model
```

```python
# discriminator
def Discriminator():
    '''
    
    '''
    dis_input = Input(shape = (64, 76, 3))
        
    model = Conv2D(filters = 64, kernel_size = 3, strides = 1, padding = "same")(dis_input)
    model = LeakyReLU(alpha = 0.2)(model)
        
    model = discriminator_block(model, 64, 3, 2)
    model = discriminator_block(model, 128, 3, 1)
    model = discriminator_block(model, 128, 3, 2)
    model = discriminator_block(model, 256, 3, 1)
    model = discriminator_block(model, 256, 3, 2)
    model = discriminator_block(model, 512, 3, 1)
    model = discriminator_block(model, 512, 3, 2)
        
    model = Flatten()(model)
    model = Dense(1024)(model)
    model = LeakyReLU(alpha = 0.2)(model)
       
    model = Dense(1)(model)
    model = Activation('sigmoid')(model) 
        
    discriminator_model = Model(inputs = dis_input, outputs = model)
    return discriminator_model
```

```python
# psnr metrics
def psnr(y_true, y_pred):
    '''
    
    '''
    return -10.0 * K.log(K.mean(K.square(y_pred - y_true))) / K.log(10.0) 
```

```python
# gan
def Gan(generator, discriminator):
    '''
    
    '''
    discriminator.trainable = False
    gan_input = Input(shape=(32, 38, 3))
    x = generator(gan_input)
    gan_output = discriminator(x)
    gan = Model(inputs=gan_input, outputs=[x,gan_output])
    gan.compile(loss=['mse', "binary_crossentropy"], loss_weights=[1., 1e-3], optimizer=Adam(lr=1e-4))

    return gan
```

```python
def train(epochs, batch_size, num_example, lr_image, hr_image):
    '''
    
    '''
    batch_count = int(num_example / batch_size)
    
    generator = Generator()
    discriminator = Discriminator()

    generator.compile(loss='mse', optimizer=Adam(lr=1e-4), metrics=['mse',psnr])
    discriminator.compile(loss="binary_crossentropy", optimizer=Adam(lr=1e-4), metrics=['accuracy'])
    
    gan = Gan(generator, discriminator)
    
    lr_image=lr_image[:num_example]
    hr_image=hr_image[:num_example]

    for e in range(1, epochs+1):
        print ('-'*15, 'Epoch %d' % e, '-'*15)
        for batch in range(batch_count):
            
            lr_image_batch = lr_image[batch*batch_size: (batch+1)*batch_size]
            hr_image_batch = hr_image[batch*batch_size: (batch+1)*batch_size]

            hr_image_gen = generator.predict(lr_image_batch)

            real_Y = np.ones(len(lr_image_batch))
            fake_Y = np.zeros(len(lr_image_batch))
            
            discriminator.trainable = True
            
            d_loss_real = discriminator.train_on_batch(hr_image_batch, real_Y)
            d_loss_fake = discriminator.train_on_batch(hr_image_gen, fake_Y)
            discriminator_loss = 0.5 * np.add(d_loss_fake, d_loss_real)

            discriminator.trainable = False
            
            gan_loss = gan.train_on_batch(lr_image_batch, [hr_image_batch,real_Y])
            
            
        print(discriminator_loss)
        print(gan_loss[0],gan_loss[1],gan_loss[2])
    return generater, discriminator, gan
```
